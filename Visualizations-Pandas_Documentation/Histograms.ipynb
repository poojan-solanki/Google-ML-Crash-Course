## [Histograms](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html#histograms)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(123456)
df = pd.DataFrame(
    {
        'a': np.random.rand(1000) + 2,
        'b': np.random.rand(1000),
        'c': np.random.rand(1000) - 2
    }
)
plt.figure()
df.plot.hist(alpha=0.8)
```

A histogram can be stacked using `stacked=True`. Bin size can be changed using the `bins` keyword.

```python
df.plot.hist(bins=20)
```

```python
df.plot.hist(bins=20, stacked=True)
```

You can pass other keywords supported by matplotlib `hist`. For example, horizontal and cumulative histograms can be drawn by `orientation='horizontal'` and `cumulative=True`.

```python
plt.figure()
```

```python
df.plot.hist(orientation='horizontal', bins=100)
```

```python
df.plot.hist(orientation='horizontal', cumulative=True)
```

See the [`hist`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html#matplotlib.axes.Axes.hist) method and the     [matplotlib hist documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) for more.

The existing interface `DataFrame.hist` to plot histogram still can be used.

```python
plt.figure()
df['a'].diff().hist()
```

[`DataFrame.hist()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.hist.html#pandas.DataFrame.hist) plots the histograms of the columns on multiple subplots:

```python
plt.figure()
df.diff().hist(color='k', bins=50, alpha=0.5)
```

```python
data = {'A': [10, 12, 15, 11],
        'B': [20, 18, 25, 22]}
data_frame = pd.DataFrame(data)

# Calculate the difference between consecutive rows
df_diff_rows = data_frame.diff()
df_diff_rows
```

The `by` keyword can be specified to plot grouped histograms:

```python
np.random.seed(123456)
data = pd.Series(np.random.randn(1000))
data.hist(by=np.random.randint(0,4,1000), figsize=(6,4)) # 1000 int numbers between 0 and 4, i.e., 0,1,2,3 
```
